```python
df_sem_nulos = df.dropna(axis=0, how='any') #remove linhas com qualquer valores nulos
df_sem_nulos
```

```python
df_drop_all = df.dropna(axis=0, how='all') #só remove se todos os valores da linha forem ausentes
df_drop_all
```

```python
df_sem_nulos_colunas = (df.dropna(axis=1))
df_sem_nulos_colunas #remove colunas que contenham valores nulos
```

```python
df_preenchido = df.fillna(0)
df_preenchido
```

```python
df_substituido = df.replace('0', np.nan) #Substitui todos os valores 0 por NaN (pode ser outros valores tambem)
df_substituido #para substituir, o 0 tem que ser passado como uma string pq é como ele está no arquivo
```

```python
#Normalizacao é o processo de ajustar os valores das colunas para um mesmo intervalo, geralmente entre 0 e 1
df['Coluna1_normalizada'] = ((df['Coluna1']) - (df['Coluna1'].mean()) / (df['Coluna1'].std()))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Hora UTC'], df['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'], marker='o', linestyle='-',color='b')
plt.xlabel('Hora UTC')
plt.ylabel('Temperatura (°C)')
plt.title('Temperatura ao Longo do Tempo')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df_20 = df.head(20)

plt.figure(figsize=(12,6))
plt.plot(df_20['Hora UTC'], df_20['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'], marker='o', linestyle='-',color='b')
plt.xlabel('Hora UTC')
plt.ylabel('Temperatura (°C)')
plt.title('Temperatura ao Longo do Tempo')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df_20['Hora UTC'], df_20['UMIDADE RELATIVA DO AR, HORARIA (%)'], marker='o', linestyle='-',color='g')
plt.xlabel('Hora UTC')
plt.ylabel('Umidade (°C)')
plt.title('Umidade ao Longo do Tempo')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Gráfico de Dispersão

plt.figure(figsize=(12,6))
plt.scatter(df_20['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'], df_20['UMIDADE RELATIVA DO AR, HORARIA (%)'], color='y')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Umidade (%)')
plt.title('Temperatura x Umidade')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(df_20['Hora UTC'], df_20['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'], color='r')
plt.xlabel('Hora UTC')
plt.ylabel('Precipitação (mm)')
plt.title('Precipitação ao Longo do Tempo')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.hist(df_20['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'], bins=10, color='#999')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Frequência')
plt.title('Histograma da temperatura')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
