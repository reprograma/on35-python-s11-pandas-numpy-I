```python
import pandas as pd

df = pd.read_csv('BRASILIA.CSV', delimiter=';', skiprows=8, encoding='latin1')
```

```python
df_amostra = df.sample(n=20)
```

```python
df_amostra.head()
```

```python
df_amostra = df_amostra[['Data', 'Hora UTC', 'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)', 'UMIDADE RELATIVA DO AR, HORARIA (%)', 'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)']]
```

```python
df_amostra
```

```python
df_amostra['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'] = df_amostra['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'].replace(',', '.', regex=True)
```

```python
df_amostra['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'] = df_amostra['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'].replace(',', '.', regex=True)
```

```python
df_amostra.head()
```

```python
df_amostra.dtypes
```

```python
df_amostra['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'] = pd.to_numeric(df_amostra['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'], errors='coerce')
```

```python
df_amostra['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'] = pd.to_numeric(df_amostra['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'], errors='coerce')
```
