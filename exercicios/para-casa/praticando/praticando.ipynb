```python
import pandas as pd
import numpy as np

df_cdba = pd.read_csv('importar/CARTEIRADIARIA_CDBA_1408.csv', delimiter=';', encoding='latin1')
df_cdneos = pd.read_csv('importar/CARTEIRADIARIA_CDNEOS_1408.csv', delimiter=';', encoding='latin1')
df_cdpe = pd.read_csv('importar/CARTEIRADIARIA_CDPE_1408.csv', delimiter=';', encoding='latin1')
df_cdrn = pd.read_csv('importar/CARTEIRADIARIA_RN_1408.csv', delimiter=';', encoding='latin1')
```

```python

df_cdba = (df_cdba[['Cliente','Código','Cotação','Quantidade Total']].dropna(axis=0))
df_cdneos = (df_cdneos[['Cliente','Código','Cotação','Quantidade Total']].dropna(axis=0))
df_cdpe = (df_cdpe[['Cliente','Código','Cotação','Quantidade Total']].dropna(axis=0))
df_cdrn = (df_cdrn[['Cliente','Código','Cotação','Quantidade Total']].dropna(axis=0))
```

```python
df = pd.concat([df_cdba,df_cdneos,df_cdpe,df_cdrn]).reset_index(drop='True') #axis = 0 -> linhas
df
```
