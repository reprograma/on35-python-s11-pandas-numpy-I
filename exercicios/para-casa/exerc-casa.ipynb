```python
import pandas as pd
import numpy as np
```

```python
#ler o arquivo
df= pd.read_csv('recife-2021.CSV', delimiter=';', skiprows=8, encoding='latin1')
```

```python
#Retirar nulos da coluna RADIACAO GLOBAL
df_reduzido = df.dropna(subset=['RADIACAO GLOBAL (Kj/m²)'])
df_reduzido
```

```python
#Quantidade de nulos antes
nulos_antes = df['RADIACAO GLOBAL (Kj/m²)'].isnull().sum()
print("Quantidade de nulos antes da remoçao: ",nulos_antes)
```

```python
#Quantidade de nulos depois da remoçao
nulos_depois = df_reduzido['RADIACAO GLOBAL (Kj/m²)'].isnull().sum()
print("Quantidade de nulos depois da remoçao: ",nulos_depois)
```

```python
#reduzindo para 3 colunas e 1000 linhas (aleatórias) 
df_reduzido = df[['Data', 'TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)', 'RADIACAO GLOBAL (Kj/m²)' ]].sample(n=1000)
df_reduzido.head(1000)
```

```python
#Conferindo se a coluna só tem valor numérico
df_reduzido['TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)'].dtype
```

```python
#Convertendo toda para valores númericos
df_reduzido = pd.to_numeric(df_reduzido['TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)'], errors='coerce')
```

```python
#Média da coluna 'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'
temperatura = df_reduzido['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)']
df_temperatura_media = round(temperatura.mean())
df_temperatura_media
```
