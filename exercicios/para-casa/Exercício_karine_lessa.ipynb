Exercício: Calcular a média da temperatura da amostra:

```python

# Transformando a coluna em float 
# calculando a média da temperatura 

coluna_temperatura_maxima = pd.to_numeric(df['TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)'], errors='coerce')
coluna_temperatura_maxima.mean()
```

Exercício: Retirar os nulos da coluna Radiação Global (Kj/m²)
