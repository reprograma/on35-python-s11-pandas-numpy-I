```python
#Importando a biblioteca pandas 
#Leitura do arquivo CSV e converter o arquivo em DataFrame
import pandas as pd
df = pd.read_csv('exercicio-para-casa-clima-SP.csv', delimiter=';', skiprows=8, encoding='latin1')
```

```python
#Importando a biblioteca pandas e numpy 
import pandas as pd
import numpy as np
```

```python
#para converter os valores da coluna para um tipo numﾃｩrico
#substituir valores desconhecidos ou invﾃ｡lidos por 'Nan" 
df['TEMPERATURA DO AR - BULBO SECO, HORARIA (ﾂｰC)'] = pd.to_numeric(df['TEMPERATURA DO AR - BULBO SECO, HORARIA (ﾂｰC)'], errors= 'coerce')
```

```python
#Mﾃｩdia da Amostra (TEMPERATURA DO AR - BULBO SECO, HORARIA (ﾂｰC)
media = df['TEMPERATURA DO AR - BULBO SECO, HORARIA (ﾂｰC)'].mean()
print(f"Mﾃｩdia: {media}")
```

```python
#verificar a quantidade de valores nulos da coluna 'Radiacao global'.
df_quantidade_nulos_radiacao = df['RADIACAO GLOBAL (Kj/mﾂｲ)'].isnull().sum()
print (f'Quantidade de valores nulos da coluna "RADIACAO GLOBAL (Kj/mﾂｲ)":{df_quantidade_nulos_radiacao}')

#Remover valores nulos coluna'Radiaﾃｧﾃ｣o Global'(Kj/m2)
df_radiacao_sem_nulos_colunas = df.dropna(subset=['RADIACAO GLOBAL (Kj/mﾂｲ)'])

#Verificar a quantidade de valores nulas da coluna'Radiaﾃｧﾃ｣o Global'(Kj/m2), apﾃｳs a remoﾃｧﾃ｣o
df_quantidade_nulos_atual = df_radiacao_sem_nulos_colunas['RADIACAO GLOBAL (Kj/mﾂｲ)'].isnull().sum()
print (f'Quantidade de valores nulos da coluna "RADIACAO GLOBAL (Kj/mﾂｲ)": {df_quantidade_nulos_atual}')
```

```python
#Copiar o dataframe e reduzir para 3 colunas e 1000 linhas (aleatﾃｳrias)
#Imprimir o DataFrame 
df_tres_colunas = df[['PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)', 'UMIDADE RELATIVA DO AR, HORARIA (%)', 'VENTO, VELOCIDADE HORARIA (m/s)']].sample(1000)
df_tres_colunas
```

```python
#Atividade Bﾃｴnus I: normalizar coluna 'TEMPERATURA Mﾃ康IMA NA HORA ANT. (AUT) (ﾂｰC)'
#Normalizar o NaN
df['TEMPERATURA Mﾃ康IMA NA HORA ANT. (AUT) (ﾂｰC)'] = pd.to_numeric(df['TEMPERATURA Mﾃ康IMA NA HORA ANT. (AUT) (ﾂｰC)'], errors='coerce')
df['TEMPERATURA Mﾃ康IMA NA HORA ANT. (AUT) (ﾂｰC)'] = df['TEMPERATURA Mﾃ康IMA NA HORA ANT. (AUT) (ﾂｰC)'].fillna(df['TEMPERATURA Mﾃ康IMA NA HORA ANT. (AUT) (ﾂｰC)'].mean())

df['coluna_normalizada'] = (df['TEMPERATURA Mﾃ康IMA NA HORA ANT. (AUT) (ﾂｰC)'] - df['TEMPERATURA Mﾃ康IMA NA HORA ANT. (AUT) (ﾂｰC)'].min()) / (df['TEMPERATURA Mﾃ康IMA NA HORA ANT. (AUT) (ﾂｰC)'].max() - df['TEMPERATURA Mﾃ康IMA NA HORA ANT. (AUT) (ﾂｰC)'].min())
```

```python
#Bﾃｴnus II - Pesquisar outras formas de processamento de dados alﾃｩm das vistas em sala de aula
#Ordenar um DataFrame pela coluna Tabela em ordem crescente e decrescente:

Tabela = {
    'Tabela': ['Data', 'Hora UTC','PRECIPITAﾃﾃグ TOTAL, HORﾃヽIO (mm)','PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)','RADIACAO GLOBAL (Kj/mﾂｲ)', 'TEMPERATURA DO AR - BULBO SECO, HORARIA (ﾂｰC)', 'UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)', 'VENTO, DIREﾃﾃグ HORARIA (gr) (ﾂｰ(gr))']
}
df = pd.DataFrame(Tabela)

# OrdenAR Dados pela coluna 'Tabela' em ordem crescente
df_ordenado_crescente = df.sort_values(by='Tabela')
print(df_ordenado_crescente)


# Ordenar dados pela coluna 'Tabela' em ordem decrescente
df_ordenado_decrescente = df.sort_values(by='Tabela', ascending=False)
print(df_ordenado_decrescente)
```

```python
##Bﾃｴnus II - Pesquisar outras formas de processamento de dados alﾃｩm das vistas em sala de aula
#Imprimir os dados de forma organizada por coluna
df = pd.DataFrame(Tabela)

#Tabela 'Ordem Crescente"
df_ordenado_crescente = df.sort_values(by='Tabela')
print("Ordem Crescente:")
print(df_ordenado_crescente.to_string(index=False)) 

#Tabela'Ordem Decrescente'
df_ordenado_decrescente = df.sort_values(by='Tabela', ascending=False)
print("\nOrdem Decrescente:")
print(df_ordenado_decrescente.to_string(index=False))
```
